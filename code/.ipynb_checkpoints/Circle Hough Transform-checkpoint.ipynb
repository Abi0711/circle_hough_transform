## Circle Hough Transform


```python
import numpy as np
import imageio.v2 as imageio
import matplotlib.pyplot as plt
import cv2 as cv
import scipy
from PIL import Image
import random
import time
```

```python
smarties_img = imageio.imread('images/smarties.png')
coins_img = imageio.imread('images/coins.png')
# Grayscale the images and cast them to integers
smarties_gray = np.uint8(np.dot(smarties_img[:,:,:3], [0.299, 0.587, 0.114]))
coins_gray = np.uint8(np.dot(coins_img[:,:,:3], [0.299, 0.587, 0.114]))
```

```python
# Use Canny Edge Detector and choose parameters that result in having the least amount of 'wispy' edges
# second and third parameter is the minvalue and maxvalue
# https://docs.opencv.org/3.4/da/d22/tutorial_py_canny.html
smarties_canny = cv.Canny(smarties_gray,300,650)
plt.figure()
plt.title("Canny Edge Detector on Smarties")
plt.imshow(smarties_canny, cmap = 'gray')
plt.figure()
plt.imshow(smarties_gray, cmap = 'gray')

coins_canny = cv.Canny(coins_gray,300,700)
plt.figure()
plt.title("Canny Edge Detector on Coins")
plt.imshow(coins_canny, cmap = 'gray')
plt.figure()
plt.imshow(coins_gray, cmap = 'gray')
plt.show()
```

```python
"""
Finds all circles in img given the following parameters
img - image that we will apply the hough circle transform on and has been previously processed by a Canny edge detector
radius_min - the minimum radius we will search for (inclusive)
radius_max - the maximum radius we will search for (exclusive)
radius_step - each step in the range between radius_min and radius_max
theta - theta values (assuming given in radians)
thresh - threshold of the accumulator
"""
def hough_circle_transform(img, radius_min,radius_max, radius_step, theta, thresh):
    (m,n) = img.shape
    # Find coordinates of all edges in the image
    edges = np.argwhere(img)
    x = edges[:,0]
    y = edges[:,1]
    r = np.arange(radius_min, radius_max,radius_step)
    # Create accumulator for a,b,r
    A = np.zeros((m, n, len(r)))
    # Precalculate degrees
    deg_a = np.cos(theta)
    deg_b = np.sin(theta)
    # Use broadcasting for the x,y coordinates for a faster result
    for radius in r:
        for t in range(len(theta)):
            # get all a and b candidates
            a_cand = np.round((x[:, np.newaxis] -(radius * deg_a[t])).flatten()).astype(int)
            b_cand = np.round((y[:, np.newaxis] - (radius * deg_b[t])).flatten()).astype(int)
            # Find candidates that are out of bounds
            bad = np.where((a_cand <= 0) | (a_cand >= m))
            bad = np.append(bad, np.where((b_cand <= 0) | (b_cand >= n)))
            # Delete them from list
            new_a = np.delete(a_cand, bad)
            new_b = np.delete(b_cand, bad)
            A[new_a, new_b, round((radius - radius_min)/radius_step)] += 1; #voting
    
    # Apply thresholding
    max_vote = np.amax(A)
    # Ratio of max votes
    A[A<thresh*max_vote] = 0
    
    # Sort the votes in descending order
    # Gets index of all non-zero which in turn gives us how many values are above the threshold
    num = len(np.argwhere(A))
    # Sort array and get the original indices 
    # unravel_index gives you the original index 
    # -A to give it in descending order
    # https://stackoverflow.com/questions/16486252/is-it-possible-to-use-argsort-in-descending-order
    # https://numpy.org/doc/stable/reference/generated/numpy.argsort.html
    sorted_indices = np.unravel_index(np.argsort(-A, axis=None), A.shape)
    # Get all the candidates in a list of form (a,b,r)
    candidates = np.array(list(zip(sorted_indices[0],sorted_indices[1], sorted_indices[2]))[:num])
    # Get actual radius value from indices
    candidates[:,2] = candidates[:,2] * radius_step + radius_min
    return candidates
```

```python
"""
Non-Maximum Suppression of circle candidates given a threshold

candidates: All circle candidates, assumption that they are in descending order of votes. Are of the form (a,b,r)
thresh: threshold for absolute difference comparison
"""
def non_maximum_suppression(candidates, thresh):
    final_candidates = []
    #compare circles with eachother
    for (a, b, r) in candidates:
        # check difference between candidate and all circles in the final candidates are greater than the threshold
        if all(abs(a - ac) > thresh or abs(b - bc) > thresh or abs(r - rc) > thresh for ac, bc,rc in final_candidates):
            final_candidates.append((a,b,r))
    return final_candidates
```

```python
# Define optimal parameters
min_rad = 20
max_rad = 30
rad_step = 1
min_theta = 0
max_theta = 360
step_theta = 3.6
thresh = 0.4
# Apply hough circle to the coins image
circles = hough_circle_transform(coins_canny, min_rad, max_rad, rad_step, (np.arange(min_theta,max_theta,step_theta)*np.pi/180) ,thresh)

# Draw circles
# https://www.geeksforgeeks.org/python-opencv-cv2-circle-method/
coin_circles = coins_img.copy()
for (a, b, r) in circles:
    coin_circles = cv.circle(coin_circles, (b,a), r,(255, 0, 0), 1)
plt.figure()
plt.title("Hough Circle Without NMS")
plt.imshow(coin_circles)

# Apply NMS
circle_final = non_maximum_suppression(circles, 7)

coin_circles = coins_img.copy()
for (a, b, r) in circle_final:
    coin_circles = cv.circle(coin_circles, (b,a), r,(255, 0, 0), 1)
plt.figure()
plt.title("Hough Circle With NMS")
plt.imshow(coin_circles)
```
